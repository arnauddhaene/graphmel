```
import numpy as np
import pandas as pd
import plotly.express as px

from scipy.stats import wasserstein_distance
```

```
mu_source, sd_source = 2.5, 0.75
mu_target, sd_target = 4.0, 1.75
```

```
ds = np.random.normal(loc=mu_source, scale=sd_source, size=1000)
dt = np.random.normal(loc=mu_target, scale=sd_target, size=1000)
```

```
df = pd.DataFrame(
    np.vstack(
       (np.concatenate((ds, dt)),
        np.concatenate(
            (np.array(['source'] * ds.shape[0]), 
             np.array(['target'] * ds.shape[0]))))
    ).T, columns=['value', 'distribution'])

df['value'] = df.value.astype(np.float32)

px.histogram(df, x='value', color='distribution') \
    .update_layout(barmode='overlay') \
    .update_traces(opacity=0.5)
```

```
wasserstein_distance(u_values=ds, v_values=dt)
```
